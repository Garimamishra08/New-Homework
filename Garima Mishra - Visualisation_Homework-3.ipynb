```python
# Import necessary libraries
import pandas as pd
import seaborn as sns

# Load the Diamonds dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Calculate mean of the price by cut
mean_price_by_cut = diamonds_data.groupby('cut')['price'].mean()
print("Mean of price by cut:")
print(mean_price_by_cut)

# Calculate standard deviation of the price by cut
std_dev_price_by_cut = diamonds_data.groupby('cut')['price'].std()
print("\nStandard Deviation of price by cut:")
print(std_dev_price_by_cut)

# Calculate median of the price by cut
median_price_by_cut = diamonds_data.groupby('cut')['price'].median()
print("\nMedian of price by cut:")
print(median_price_by_cut)

# Calculate mode of the price by cut
mode_price_by_cut = diamonds_data.groupby('cut')['price'].apply(lambda x: x.mode().iloc[0])
print("\nMode of price by cut:")
print(mode_price_by_cut)

# Calculate variance of the price by cut
variance_price_by_cut = diamonds_data.groupby('cut')['price'].var()
print("\nVariance of price by cut:")
print(variance_price_by_cut)

# Calculate 25th and 75th percentile of the price by cut
percentiles_25_75_price_by_cut = diamonds_data.groupby('cut')['price'].quantile([0.25, 0.75])
print("\n25th and 75th Percentiles of price by cut:")
print(percentiles_25_75_price_by_cut)

# Calculate range of the price by cut (max - min)
price_range_by_cut = diamonds_data.groupby('cut')['price'].max() - diamonds_data.groupby('cut')['price'].min()
print("\nRange of price by cut:")
print(price_range_by_cut)
```

```python
import pandas as pd
import seaborn as sns

# Load the Diamonds dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Calculate correlation between price and carat
correlation_price_carat = diamonds_data['price'].corr(diamonds_data['carat'])
print("Correlation between price and carat:")
print(correlation_price_carat)

# Calculate covariance between price and carat
covariance_price_carat = diamonds_data['price'].cov(diamonds_data['carat'])
print("\nCovariance between price and carat:")
print(covariance_price_carat)
```

```python
# 2. Barplot: Average Diamond Price by Color
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Diamonds dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Calculate the average price by color
avg_price_by_color = diamonds_data.groupby('color')['price'].mean().reset_index()

# Set up the plot using seaborn
plt.figure(figsize=(10, 6))  # Set the figure size (width, height)

# Create a barplot using seaborn
sns.barplot(x='color', y='price', data=avg_price_by_color, palette='coolwarm')

# Set the plot title and labels
plt.title('Average Diamond Price by Color')
plt.xlabel('Color')  # X-axis label
plt.ylabel('Average Price')  # Y-axis label

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Diamonds dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Set up the scatterplot with facets using seaborn
plt.figure(figsize=(12, 8))  # Set the overall figure size (width, height)

# Create a scatterplot with facets using seaborn
sns.relplot(x='carat', y='price', hue='cut', data=diamonds_data, col='cut', col_wrap=3, height=4, aspect=1.2, palette='viridis')

# Set the plot title and labels
plt.suptitle('Scatterplot of Carat vs. Price by Cut', y=1.02)  # Title at the top of the figure
plt.xlabel('Carat')  # X-axis label for each subplot
plt.ylabel('Price')  # Y-axis label for each subplot

# Show the plot
plt.show()
```

```python
# 4. Pie Chart: Distribution of Diamond Cut
```

```python
# Load the Diamonds dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Calculate the distribution of diamond cuts
cut_distribution = diamonds_data['cut'].value_counts()

# Set up the pie chart using matplotlib
plt.figure(figsize=(8, 8))  # Set the figure size (width, height)

# Create a pie chart
plt.pie(cut_distribution, labels=cut_distribution.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)

# Set the plot title
plt.title('Distribution of Diamond Cut')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Diamonds dataset from Seaborn
diamonds_data = sns.load_dataset("diamonds")

# Select numeric columns for correlation analysis
numeric_columns = diamonds_data.select_dtypes(include=['number'])

# Create a correlation matrix
correlation_matrix = numeric_columns.corr()

# Set up the correlation plot (heatmap) using seaborn
plt.figure(figsize=(10, 8))  # Set the figure size (width, height)

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Set the plot title
plt.title('Correlation Matrix of Diamond Attributes')

# Show the plot
plt.show()
```
